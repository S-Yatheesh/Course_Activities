2 . Create two 3×3 matrices:

One filled with all 5s , One with values from 1 to 9

```python
matrix_1 = np.full((3,3),5)
print(f"First Matrix : \n",matrix_1)
matrix_2 = np.arange(1,10)
matrix_2 = matrix_2.reshape(3,3)
print(f"Second Matrix :\n",matrix_2)
```


● Add the two matrices

```python
sum_of_matrix = np.add(matrix_1,matrix_2)
print(sum_of_matrix)
```


● Multiply them element-wise

```python
Product_of_matrix = np.multiply(matrix_1,matrix_2)
print(Product_of_matrix)
```


● Then find the transpose of the result


```python
Transpose = np.transpose(Product_of_matrix)
print(Transpose)
```


4.Generate a 5×4 array with random values between 0 and 1 (float).

```python
arr_4 = np.random.rand(5, 4)
arr_4
```

● Find the row with the maximum sum of values.

```python
row_sums = arr_4.sum(axis=1)
max_row_index = np.argmax(row_sums)
print(f"The row with the maximum sum is:\n{arr_4[max_row_index]}")
```

● Normalize each row using the formula:

(x-min)/(max-min)

```python
normalized = (arr_4 - arr_4.min(axis=1, keepdims=True)) / (arr_4.max(axis=1, keepdims=True) - arr_4.min(axis=1, keepdims=True))
print("Normalized Array:\n", normalized)
```

5.arr = np.array([1, 3, 5, 3, 7, 9, 1, 5, 5])

● Find all unique elements

● Find the duplicates (elements that appear more than once)


```python
arr = np.array([1, 3, 5, 3, 7, 9, 1, 5, 5])
unique_elements = np.unique(arr)

vals, counts = np.unique(arr, return_counts=True)
duplicates = vals[counts > 1]

print("Unique:", unique_elements)
print("Duplicates:", duplicates)
```

● Join them into a single array


```python
a, b, c = np.array([1, 2, 3]), np.array([4, 5, 6]), np.array([7, 8, 9])
joined = np.concatenate([a, b, c])
print(joined)
```


● Reshape the result into a 3×3 matrix



```python
result = joined.reshape(3, 3)
print("Matrix:\n", result)
```

● Extract the diagonal elements

```python
diagonal = np.diag(result)
print("Diagonal elements:", diagonal)
```
